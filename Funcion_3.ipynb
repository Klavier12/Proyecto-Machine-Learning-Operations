```python
import pandas as pd 
import numpy as np 
```

```python
#importamos los datasets previamente transformados
#Primero ponemos la ruta
ruta_games = r"C:\Users\Cristian\Desktop\DATA SCIENCE\HENRY\PI1\PI MLOps - STEAM\Datos Estructurados\games.parquet"
ruta_items = r"C:\Users\Cristian\Desktop\DATA SCIENCE\HENRY\PI1\PI MLOps - STEAM\Datos Estructurados\users_items.parquet"
ruta_reviews = r"C:\Users\Cristian\Desktop\DATA SCIENCE\HENRY\PI1\PI MLOps - STEAM\Datos Estructurados\df_review_final.parquet"
#luego los extraemos
games = pd.read_parquet(ruta_games)
items = pd.read_parquet(ruta_items)
#reviews = pd.read_parquet(ruta_reviews)
```

```python
#Necesitamos el dataset games pero no todas sus columnas
games_col = games.drop(columns=["Price","Developer", "Publisher"])
#Pasamos esta columna a tipo date time para obtener el año
games_col["Release_Date"] = pd.to_datetime(games_col["Release_Date"], errors = "coerce")
#Eliminamos los nulos
games_col = games_col.dropna(subset=["Release_Date"])
#Obtenemos el año
games_col["Year"] = games_col["Release_Date"].dt.year
#Eliminamos la columna ya que no es necesaria
games_col.drop(columns=["Release_Date"], inplace=True)
games.rename(columns={"Id": "Item_Id"}, inplace = True)
games_col.head(2)
```

```python
# Examinamos los datasets que necesitamos
items.columns = items.columns.str.title()
#Extraemos las columnas que necesitamos
items_col = items[["User_Id", "Item_Id", "Item_Name", "Playtime_Forever"]]
items_col.head(5)
```

```python
# Vamos a revisar el tipo de dato de estas columnas ya que son las que usaremos para armar un solo dataset
print(items_col["Item_Id"].dtype)
print(games_col["Item_Id"].dtype)
```

```python
# Vamos a crear un nuevo dataset con las columnas necesarias
# Cambiamos tipo de datos para union
items_col["Item_Id"] = items_col["Item_Id"].astype("int32")
# Ahora procedemos a unir los dos datsets
funcion_3 = games_col.merge(items_col, on="Item_Id", how="inner")
funcion_3
```

```python
# Eliminamos la columna Item_Name ya que tiene los mismos datos que App_Name
funcion_3.drop(columns=["Item_Name"], inplace=True)
funcion_3.head(5)
```

```python
# Ahora cambiamos el orden de las columnas para mejorar su visibilidad
funcion_3 = funcion_3[["User_Id", "App_Name", "Item_Id", "Genres", "Playtime_Forever", "Year"]]
funcion_3.head(5)
```

```python
# Eliminamos datos nulos
funcion_3.dropna(inplace = True)
```

```python
# Debido a la gran cantidad de datos solo voy a tener en cuenta desde el año 2014
funcion_3 = funcion_3[funcion_3["Year"] > 2007]
# Vemos el Dataset
funcion_3.info()


```

```python
#def UserForGenre( genero : str ): Debe devolver el usuario que acumula más horas jugadas para el género dado y una lista de la acumulación de horas jugadas por año de lanzamiento.
#Ejemplo de retorno: {"Usuario con más horas jugadas para Género X" : us213ndjss09sdf, "Horas jugadas":[{Año: 2013, Horas: 203}, {Año: 2012, Horas: 100}, {Año: 2011, Horas: 23}]}
def UserForGenre(genero: str):
    '''
    Esta función devuelve información sobre el usuario con más horas de juego según el género de sus items.

    Args:
        genero (str): Nombre del género.

    Returns:
        dict: Un diccionario que contiene información sobre el usuario con más horas de juego por año y género.
            - 'User Id' (int): Identificador único del usuario.
            - 'Year' (int): Año.
            - 'Playtime_Forever' (float): Número de horas jugadas.
    '''    
    # Explode los géneros en el DataFrame
    df_exploded = funcion_3.explode('Genres')
    
    # Filtramos por el género dado
    genre_filtered = df_exploded[df_exploded['Genres'].apply(lambda x: genero in x)]
    
    if genre_filtered.empty:
        return f"No hay datos para el género {genero}"

    # Agrupamos por User_Id y Year, y acumulamos horas jugadas
    hours_per_year = genre_filtered.groupby(['User_Id', 'Year'])['Playtime_Forever'].sum().reset_index()

    # Sumamos horas totales por usuario
    total_hours = hours_per_year.groupby('User_Id')['Playtime_Forever'].sum().reset_index()

    # Encontramos el usuario con más horas
    top_user = total_hours.loc[total_hours['Playtime_Forever'].idxmax()]

    # Creamos la lista de horas por año
    hours_by_year = hours_per_year[hours_per_year['User_Id'] == top_user['User_Id']].to_dict(orient='records')

    return {
        "Usuario con más horas jugadas para Género": top_user['User_Id'],
        "Horas jugadas": hours_by_year
    }


```
