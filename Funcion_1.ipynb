```python
#importamos los datasets previamente transformados
#Primero ponemos la ruta
ruta_games = r"C:\Users\Cristian\Desktop\DATA SCIENCE\HENRY\PI1\PI MLOps - STEAM\Datos Estructurados\games.parquet"
ruta_items = r"C:\Users\Cristian\Desktop\DATA SCIENCE\HENRY\PI1\PI MLOps - STEAM\Datos Estructurados\users_items.parquet"
ruta_reviews = r"C:\Users\Cristian\Desktop\DATA SCIENCE\HENRY\PI1\PI MLOps - STEAM\Datos Estructurados\df_review_final.parquet"
#luego los extraemos
games = pd.read_parquet(ruta_games)
#items = pd.read_parquet(ruta_items)
reviews = pd.read_parquet(ruta_reviews)
```

```python
#Funcion 1: developer()
#def developer( desarrollador : str ): Cantidad de items y porcentaje de contenido Free por año según empresa desarrolladora. Ejemplo de retorno:
funcion_1 = games.copy()
funcion_1.drop(columns=["Publisher", "Genres"], inplace = True) #Eliminamos columnas que no sirven

funcion_1.head(5)
```

```python
#Ahora procedemos a crear la columna Year
funcion_1['Release_Date'] = pd.to_datetime(funcion_1['Release_Date'], errors='coerce')
funcion_1["Year"] = funcion_1['Release_Date'].dt.year
funcion_1.drop(columns=["Release_Date"], inplace = True)
funcion_1.head(4)
```

```python
funcion_1["Developer"].value_counts()
```

```python
def developer(desarrollador: str):
    '''
    Esta funcion devuelve la cantidad de items y porcentaje de contenido Free por año según empresa desarrolladora. 
    
    Args : desarrollador (str) Nombre del Desarrollador.

    Returns:
        Dataframe con las siguientes columnas:
            - Año: Contexto de tiempo
            - Cantidad de items: Cantidad de videojuegos
            - Contenido Free: Cantidad de juegos gratuitos
    '''
    # Primero filtrar el DataFrame por el desarrollador
    df_dev = funcion_1[funcion_1['Developer'] == desarrollador]

    # Nos aseguramos que la columna Price sea numerica
    df_dev['Price'] = pd.to_numeric(df_dev['Price'], errors='coerce')

    # Creamos la columna 'Free' donde 1 indica que el juego es gratis
    df_dev['Free'] = df_dev['Price'].apply(lambda x: 1 if x == 0 else 0)

    # Agrupamos por 'Year' para calcular cantidad de juegos y porcentaje de juegos gratuitos
    result = df_dev.groupby('Year').agg(
        Cantidad_Items=('Id', 'count'),
        Contenido_Free=('Free', lambda x: f"{(x.mean() * 100):.0f}%")
    ).reset_index()

    return result

# Testeamos
developer('Paradox Development Studio')
```
